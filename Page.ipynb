```python
from tkinter import *
from tkinter import messagebox as mbox
from tkinter import simpledialog,ttk
import pandas as pd
import seaborn as sns
from nltk.corpus import stopwords
from string import punctuation
import nltk
import cv2
from sklearn.model_selection import train_test_split
from sklearn.feature_extraction.text import CountVectorizer
from sklearn.feature_extraction.text import TfidfTransformer
from sklearn.pipeline import Pipeline
from sklearn.ensemble  import RandomForestClassifier

def process_text(mess):
    mess = [word for word in mess if word not in punctuation]
    mess = ''.join(mess)
    mess = [word for word in mess.split() if word not in stopwords.words('english')]
    return mess

def opencv():
    cam=cv2.VideoCapture(0)
    cv2.namedWindow("Python Web")
    img_counter=0
    while True:
        ret,frame=cam.read()
        if not ret:
            print("failed to grab frame")
            break
        cv2.imshow('test',frame)
        k=cv2.waitKey(1)
        if k%256==27:
            print("")
            break
        elif k%256==32:
            img_name="opencvframe{}.png".format(image_counter)
            cv2.imwrite(img_name,frame)
            img_counter+=1
    cam.release()
    cam.destroyAllWindows()
def spamwindow():
    try:
        root.destroy()
        global root2
        root2=Tk()
        root2.title('Spam')
        root2.geometry('1100x800')
        bg=PhotoImage(file='img/spam.png')
        Label(root2,image=bg).place(relwidth=1,relheight=1)
        #add opencv
        
        root2.mainloop()
        opencv()
    except:pass
def hamwindow():
    try:
        root.destroy()
        global root1
        root1=Tk()
        root1.title('Ham')
        root1.geometry('1100x800')
        bg=PhotoImage(file='img/nospam.png')
        Label(root1,image=bg).place(relwidth=1,relheight=1)
        root1.mainloop()
    except:pass
def evaluate():
    global text
    text=txt.get()
    messages = pd.read_csv('SMSSpamCollection',sep='\t',names=['label','message'])
    msg_train,msg_test,label_train,label_test = train_test_split(messages['message'],messages['label'],test_size=0.3)
    pipeline = Pipeline([('bow',CountVectorizer(analyzer=process_text)),('tfidf',TfidfTransformer()),('classifier',RandomForestClassifier())])
    pipeline.fit(msg_train,label_train)
    p=pipeline.predict([f'{text}'])[0]
    try:
        if p=='ham':hamwindow()
        else:spamwindow()
    except:pass
def main_window():
    global root,txt
    root=Tk()
    root.title('Spam Detector')
    root.geometry('1100x800')
    bg=PhotoImage(file='enter.png')
    Label(root,image=bg).place(relwidth=1,relheight=1)
    txt=StringVar()
    text=Entry(root,width=65,textvariable=txt).place(x=390,y=350)
    btpic=PhotoImage(file='sub2.png')
    b1=Button(root,text='Submit',padx=30,pady=15,bg='#000000',image=btpic,command=evaluate)
    b1.place(x=500,y=390)
    root.mainloop()

if __name__ == "__main__":
    main_window()   
```
